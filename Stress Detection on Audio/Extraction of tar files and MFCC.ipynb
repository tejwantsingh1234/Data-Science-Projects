# Stress detection using Audio data

## Data Collection

The data was collected from a database given on the below link 
https://dcapswoz.ict.usc.edu/wwwedaic/data/

The Extended Distress Analysis Interview Corpus (E-DAIC) (DeVault et al., 2014) is an
extended version of WOZ-DAIC (Gratch et al.,2014) that contains semi-clinical
interviews designed to support the diagnosis of psychological distress conditions such
as anxiety, depression, and post-traumatic stress disorder. These interviews were
collected as part of a larger effort to create a computer agent that interviews people and
identifies verbal and nonverbal indicators of mental illnesses.
The interviews are conducted by an animated virtual interviewer called Ellie.
A subset of the sessions are collected in a wizard-of-Oz (WoZ) setting, where the virtual
agent is controlled by a human interviewer (wizard) in another room.
The other subset are collected using an AI-controlled agent, who acts in a fully
autonomous way using different automated perception and behavior generation
modules.

## Extracting .tar files and removing the old ones

```python
import os
import shutil

def extract_and_remove_tar(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)
    
    # Iterate through each file
    for file in files:
        if file.endswith('.tar'):
            # Extract tar file
            tar_path = os.path.join(folder_path, file)
            shutil.unpack_archive(tar_path, folder_path)
            print(f"Extracted {tar_path}")

            # Remove tar file
            os.remove(tar_path)
            print(f"Removed {tar_path}")

folder_path = r"D:\Sehaj\Data Analyst\Projects\Stress Detection using Audio\data"
extract_and_remove_tar(folder_path)
```

## Feature Extraction

### Controlled

```python
import librosa
import matplotlib.pyplot as plt

# Define input and output folders
input_folder = r"D:\Sehaj\Data Analyst\Projects\Stress Detection using Audio\Controlled"
output_folder = r"D:\Sehaj\Data Analyst\Projects\Stress Detection using Audio\Controlled_feat"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Function to plot MFCC and save it to output folder
def plot_mfcc(audio_file_path, output_file_path):
    # Load audio file
    y, sr = librosa.load(audio_file_path)
    
    # Extract MFCC features
    mfccs = librosa.feature.mfcc(y=y, sr=sr)
    
    # Plot MFCC
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfccs, x_axis='time')
    plt.colorbar()
    plt.title('MFCC')
    plt.tight_layout()
    plt.savefig(output_file_path)
    plt.close()

# Iterate over subfolders in the input folder
for root, dirs, files in os.walk(input_folder):
    for dir_name in dirs:
        dir_path = os.path.join(root, dir_name)
        print("Processing folder:", dir_path)
        
        # Iterate over audio files with .wav extension
        for file_name in os.listdir(dir_path):
            if file_name.endswith(".wav"):
                audio_file_path = os.path.join(dir_path, file_name)
                
                # Define output file path
                output_file_name = os.path.splitext(file_name)[0] + "_mfcc.png"
                output_file_path = os.path.join(output_folder, output_file_name)
                
                # Plot MFCC and save it to output folder
                plot_mfcc(audio_file_path, output_file_path)
                
                print("Processed audio file:", file_name)
                print("MFCC plot saved to:", output_file_path)
```
